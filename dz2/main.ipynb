```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
first = pd.read_csv('first.csv', sep=' ')
second = pd.read_csv('second.csv', sep=' ')
third = pd.read_csv('third.csv', sep=' ')
forth = pd.read_csv('forth.csv', sep=' ')
fifth = pd.read_csv('fifth.csv', sep=' ')
```

```python
F_score = 0.55
```

```python
import numpy
temp = pd.merge(pd.concat([first, second], axis=0), pd.concat([third, forth], axis=0), on='Id')
temp = temp.fillna(0)
temp["Age"] = temp["Age_x"] + temp["Age_y"]
temp = temp.drop(columns=['Age_x', 'Age_y'])
temp = pd.merge(temp, fifth, on='Id')
temp.to_csv("merged.csv", sep=" ", index=False)
```

```python
data = pd.read_csv("merged.csv", sep=" ")
data.head(10)
```

```python
data['Gender'] = data['Gender'].map({'Female': 1, 'Male': 0})
featured_data = data.drop(columns=['CustomerId','Surname','Geography'])
featured_data
```

```python
from sklearn.model_selection import train_test_split
train_data, vnt_data = train_test_split(featured_data, test_size=0.40, random_state=23042020)
valid_data, test_data = train_test_split(vnt_data, test_size=0.50, random_state=23042020)
```

```python
train_data_features = train_data.drop(['Exited'], axis=1)
train_data_target = train_data['Exited']
valid_data_features = valid_data.drop(['Exited'], axis=1)
valid_data_target = valid_data['Exited']
test_data_features = test_data.drop(['Exited'], axis=1)
test_data_target = test_data['Exited']
```

Начнем с Дерева выбора. Изменим в цикле гиперпараметр max_depth.

```python
%%time

best_depth = 0
best_accuracy = 0

for depth in range(1,101):
    model = DecisionTreeClassifier(random_state=23042020, max_depth=depth)
    model.fit(train_data_features, train_data_target)
    valid_predictions = model.predict(valid_data_features)
    accuracy = accuracy_score(valid_data_target, valid_predictions)
    print('Глубина дерева', depth,'Точность',accuracy)
    if accuracy > best_accuracy:
        best_depth = depth
        best_accuracy = accuracy
        
print('Лучшая глубина дерева', best_depth,'Лучшая точность', best_accuracy)   
```

```python
%%time

best_tree_model = DecisionTreeClassifier(random_state=23042020, max_depth=3)
best_tree_model.fit(train_data_features, train_data_target)
valid_predictions = best_tree_model.predict(valid_data_features)
accuracy = accuracy_score(valid_data_target, valid_predictions)
print('Глубина дерева', 3,'Точность',accuracy)
```

Перейдем к оценке случайного леса. Изменим в цикле гиперпараметр n_estimators.

```python
%%time

best_estim = 0
best_accuracy = 0

for estim in range(1,101):
    model = RandomForestClassifier(random_state=23042020, n_estimators=estim)
    model.fit(train_data_features, train_data_target)
    valid_predictions = model.predict(valid_data_features)
    accuracy = accuracy_score(valid_data_target, valid_predictions)
    print('Количество деревьев', estim,'Точность',accuracy)
    if accuracy > best_accuracy:
        best_estim = estim
        best_accuracy = accuracy

print('Лучшая количество деревьев', best_estim,'Лучшая точность', best_accuracy)      
```

```python
%%time

best_forest_model = RandomForestClassifier(random_state=23042020, n_estimators=10)
best_forest_model.fit(train_data_features, train_data_target)
valid_predictions = best_forest_model.predict(valid_data_features)
accuracy = accuracy_score(valid_data_target, valid_predictions)
print('Количество деревьев', 8,'Точность',accuracy)
```

```python
%%time
import warnings
from sklearn.exceptions import ConvergenceWarning
warnings.filterwarnings('ignore', category=ConvergenceWarning, module='sklearn.neural_network')
best_max_iter = 0
best_accuracy = 0

for max_iter in range(1,201):
    model = LogisticRegression(random_state=23042020, max_iter=max_iter)
    model.fit(train_data_features, train_data_target)
    valid_predictions = model.predict(valid_data_features)
    accuracy = accuracy_score(valid_data_target, valid_predictions)
    print('Количество итерраций', max_iter,'Точность',accuracy)
    if accuracy > best_accuracy:
        best_max_iter = max_iter
        best_accuracy = accuracy

print('Лучшая количество итерраций', best_max_iter,'Лучшая точность', best_accuracy)
```

```python
%%time
import warnings
from sklearn.exceptions import ConvergenceWarning
warnings.filterwarnings('ignore', category=ConvergenceWarning, module='sklearn.neural_network')

best_reg_model = LogisticRegression(random_state=23042020, max_iter=168)
best_reg_model.fit(train_data_features, train_data_target)
valid_predictions = best_reg_model.predict(valid_data_features)
accuracy = accuracy_score(valid_data_target, valid_predictions)
print('Количество итерраций', 168,'Точность',accuracy)
```

```python
test_predictions = best_tree_model.predict(test_data_features)
accuracy = accuracy_score(test_data_target, test_predictions)
print('Точность модели решающего леса на тестовой выборке',accuracy)

test_predictions = best_forest_model.predict(test_data_features)
accuracy = accuracy_score(test_data_target, test_predictions)
print('Точность модели случайного леса на тестовой выборке',accuracy)

test_predictions = best_reg_model.predict(test_data_features)
accuracy = accuracy_score(test_data_target, test_predictions)
print('Точность модели логистической регресиии на тестовой выборке',accuracy)
```
